```python
from covid import Covid #Importing Module
cv=Covid()
for _ in range(100):
  print()
  print("***Covid Counter***")

  #Global count by using inbulit function
  #Assigned varibles to this function
  act=cv.get_total_active_cases()
  rec=cv.get_total_recovered()
  death=cv.get_total_deaths()
  conf=cv.get_total_confirmed_cases()

  #For getting updates
  print("1. Global Count")# get global count
  print("2. Active Cases")
  print("3. Confirmed Cases")
  print("4. Recovered Cases")
  print("5. Deceased")
  print("6. Get Covid Updates by Country Name")

  # Get choice by user
  Choice=int(input('Enter the number of choices '))

  if(Choice==1):
    print(f" Active cases: {act}\n Confiremed Cases: {conf}\n Recovered Cases: {rec}\n Deceased:{death}")
  elif(Choice==2):
    print(f"Active cases: {act}")
  elif(Choice==3):
    print(f"Confirmed Cases: {conf}")
  elif(Choice==4):
    print(f"Recovered Cases: {rec}")
  elif(Choice==5):
    print(f"Deceased: {death}")
  elif(Choice==6):
    Country_name=input("Enter the Country Name ")
    Cname=cv.get_status_by_country_name(Country_name)
    #print(Cname)
    print(f"Active Cases: {Cname['active']}")
    print(f"Confirmed Cases: {Cname['confirmed']}")
    print(f"Deceased: {Cname['deaths']}")
    print(f"Recovered Cases: {Cname['recovered']}")
    
  else:
    print("Invalid Choice")
    print()
    break
```
